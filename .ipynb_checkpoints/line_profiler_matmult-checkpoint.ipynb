```python
# Program to multiply two matrices using nested loops
import random
```

```python
%lprun -f random random
```

```python
# NxN matrix
#Wrap the code inside a function so that it can be profiled later.
X = []
def Xmatrix(N):
    for i in range(N):
        X.append([random.randint(0,100) for r in range(N)])
```

```python
#pass the N parameter to the function and then call %lprun
%lprun -f Xmatrix Xmatrix(N)
```

```python
# Nx(N+1) matrix
Y = []
def Ymatrix(N):
    for i in range(N):
        Y.append([random.randint(0,100) for r in range(N+1)])
```

```python
N = 250
%lprun -f Ymatrix Ymatrix(N)
```

```python
# result is Nx(N+1)
result = []
def Rmatrix(N):
    for i in range(N):
        result.append([0] * (N+1))
```

```python
N = 250
%lprun -f Rmatrix Rmatrix(N)
```

```python
def iterat():
    # iterate through rows of X
    for i in range(len(X)):
        # iterate through columns of Y
        for j in range(len(Y[0])):
            # iterate through rows of Y
            for k in range(len(Y)):
                result[i][j] += X[i][k] * Y[k][j]
```

```python
%lprun -f iterat iterat()
```

```python
def resfunc():
    for r in result:
        print(r)
```

```python
%lprun -f resfunc resfunc()
```
